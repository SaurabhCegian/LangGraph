```python
from typing import List, TypedDict
from langgraph.graph import StateGraph
import math
```

```python
class AgentState(TypedDict):
  name : str
  values : List[int]
  operation : str
  result : str

def process_value(state : AgentState) -> AgentState:
  state['result'] = f"Hi {state['name']} "
  if state['operation'] == '*':
    state['result'] = state['result'] + f"your result is {math.prod(state['values'])}"
  if state['operation'] == '+':
    state['result'] = state['result'] + f"your result is {sum(state['values'])}"
  return state

graph = StateGraph(AgentState)
graph.add_node("processor", process_value)
graph.set_entry_point("processor")
graph.set_finish_point("processor")

app = graph.compile()
answer = app.invoke({"name":"Saurabh", "values":[1,2,3,4], "operation":"*"})
print(answer)
```
